# Анализ коэффициента удержания в приложении покупок магазина

В июле 2015 года провели специальную акцию, направленную на улучшение ретеншена. Новым юзерам слали письмо с купоном на скидку.

## Задачи исследования

- Выяснить, сработала ли акция?
- Как повел себя коэффициент удержания после проведения акции?

## Исходные данные

- `user_id` - идентификатор пользвателя
- `order_id` - идентификатор заказа
- `reg_date` - дата регистрации пользователя
- `transaction_date` - дата оформления заказа
- `revenue` - выручка заказа

## Предобработка данных

## Исследовательский анализ

### Изучим переменную с датами регистраций `reg_date`

#### Посмотрим на период регистраций пользователей

```python
data[['reg_date']].agg(['min', 'max'])
```

Дата начала периода регистраций явно неккоректна. Посмотрим на другие значения дат в начале периода

```python
data.groupby('reg_date').agg({'user_id': 'count'}).reset_index().sort_values(by='reg_date').head()
```

Видно, что присутствует только одна неккоректная дата, которая имеется у 53 наблюдений в наших данных. Так как наблюдений немного, просто удалим их

```python
data = data[data['reg_date'] >= '2012-02-24']
```

Теперь посмотрим на период регистрации пользователей только с корректными значениями

```python
data[['reg_date']].agg(['min', 'max'])
```

```python
print(data['reg_date'].max() - data['reg_date'].min())
```

Пользователи в наших данных регистрировались начиная с 2012 и заканчивая концом 2015 года.

#### Динамика изменения количества новых пользователей по годам

```python
data['reg_year'] = data['reg_date'].astype('datetime64[Y]')
```

```python
data.groupby('reg_year').agg({'user_id': 'nunique'}).plot(figsize=(7, 5), grid=True)
plt.title('Динамика регистраций новых пользователей по годам')
plt.show()
```

- Начиная с 2014 года прирост новых пользователей несколько уменьшился, по сравнению с предыдущими годами

#### Динамика изменения количества новых пользователей по месяцам

```python
data['reg_month'] = data['reg_date'].astype('datetime64[M]')
```

```python
data.groupby('reg_month').agg({'user_id': 'nunique'}).plot(figsize=(15, 7), grid=True)
plt.title('Динамика регистраций новых пользователей по месяцам')
plt.show()
```

Прирост новых пользователей в 2015 году очень волатилен и под конец наблюдается аномально резкое падение количества регистраций

#### Относительный прирост новых пользователей по месяцам

Посмотрим на прирост новых пользователей по месяцам, относительно общего прироста за соответсвующий год

```python
data['reg_month_n'] = data['reg_date'].dt.month

users_by_years = data.groupby('reg_year').agg(
    users_by_year=('user_id', 'nunique')).reset_index()

users_by_months = (
    data.groupby(['reg_year', 'reg_month'])
        .agg(users_by_month=('user_id', 'nunique'))
        .reset_index()
)

users_by_months = users_by_months.merge(users_by_years, how='left', on='reg_year')
users_by_months['month_proportion'] = users_by_months['users_by_month'] / users_by_months['users_by_year']
```

Объединили две таблицы с данными количества регистраций по годам и месяцам и рассчитали долю регистраций в каждый месяц за соответсвующий год. Теперь построим линейный график

```python
users_by_months[['reg_month', 'month_proportion']].plot(
    x='reg_month', 
    y='month_proportion',
    figsize=(10, 7),
    grid=True
)
plt.ylabel('Доля новых пользователй')
for year in users_by_months['reg_year'].dt.year.unique():
    plt.axvspan(f'{year}-07-01', f'{year+1}-01-01', alpha=0.5, color='lightblue')
plt.title('Динамика изменения доли прироста новых пользователей по месяцам')
plt.show()
```

Закрашенные области - периоды с июля до конца года, для выявления возможных сезонных особенностей в привлечении новых пользователей

- На графике видим некую сезонность - в конце каждого года, кроме 2015, количество пользователей довольно резко увеличивается вплоть до декабря, далее идет резкое снижение
- Весь наш исследуемый период (с июля до конца 2015 года) идет резкое снижение регистрации новых пользователей

## Когортный анализ

Наша задача выяснить увеличился ли коэффициент удержания в связи с проведением акции в июле для новых пользователей. Для этого проведем когортный анализ, где когортами у нас будут пользователи зарегистрировавшиеся в каждый месяц 2015 года, а время жизни - лайфтайм будет измеряться в неделях

#### Создадим сводную таблицу с когортами и их лайфтаймом

Сделаем срез пользователей которые зарегистрировались в 2015 году, так как у нас имеются данные только для заказов сделанных в этом году

```python
data = data[data['reg_date'] >= '2015-01-01']
```

Для каждого наблюдения выделим лайфтайм в неделях

```python
data['lifetime'] = (data['transaction_date'].dt.to_period('W').astype('int') - 
                    data['reg_date'].dt.to_period('W').astype('int'))
```

```python
data.head()
```

Теперь создадим сводную таблицу, где:
- индексы - это наши когорты по месяцу регистрации в 2015 году
- колонки - это лайфтайм в неделях
- значения - это количество уникальных пользователей по когортам и на определенный лайфтайм

```python
retention = data.pivot_table(
    index='reg_month',
    columns='lifetime',
    values='user_id',
    aggfunc='nunique'
)

retention = retention.rename(columns={0: 'cohort_size'})
retention
```

#### Сделаем срез данных на основе горизонта анализа

Рассчитаем горизонт анализа в неделях. Для этого найдем разницу в неделях между максимальной датой трансакций и датой начала акции то есть 01 июля 2015

```python
horizon_weeks = int((data['transaction_date'].max() - pd.to_datetime('2015-07-01')).days / 7)
horizon_weeks
```

Получили горизонт анализа 20 недель. Теперь сделаем срез данных, где максимальная когорта будет пользователями пришедшими в июле, а максимальный лайфтайм ограничен в 20 недель

```python
retention = retention.loc[:'2015-07-01', :20]
```

Теперь рассчитаем Retention Rate для чего значения в колонках начиная с первой поделим на размеры когорт

```python
retention = retention.iloc[::, 1:].div(retention['cohort_size'], axis=0)
retention
```

#### Построим тепловую карту

```python
plt.figure(figsize=(20, 10))
sns.heatmap(retention, annot=True, cbar=False)
plt.title('Тепловая карта коэффициента удержания')
plt.yticks(ticks=range(1, 8), labels=range(1, 8))
plt.show()
```

- На графике видим, что пользователи зарегистрировавшиеся в июле (месяц проведения акции), на конец периода в 20 недель имеют самый маленький retention rate. Однако ранее на [графике динамики заказов](#orders_lineplot) мы видели очень резкое падение после достижения максимума в октябре.
- Так как это падение может быть связано с техническими причинами, то будем смотреть retention до аномального падения то есть на 16-17 неделе лайфтайма
- Но даже по лайфтайму на 16-17 неделе сказать, что проведенная акция помогла в удержании пользователей зарегистрировавшихся в июле нельзя

#### Посмотрим динамику изменения retention на линейном графике

```python
retention.T.plot(figsize=(10, 7), color=['grey']*6+['red'])
plt.title('Динамика изменения Retention Rate по когортам')
plt.legend(retention.index.astype('str'))
plt.xticks(range(1, 21))
plt.show()
```

- Наша целевая когорта лидировала по коэффициенту удержания до 7 недели лайфтайма
- Начиная с 17 недели лайфтайма идет резкое снижение коэффициента вплоть до самых низких показателей среди всех когорт, что мы выявили на предыдущем графике

## Выводы

- По результатам исследования можно предположить, что акция проведенная с целью улучшения Retention Rate не принесла положительных результатов. 
- Есть подозрения, что прсутствуют дополнительные причины, провала целевой когорты, связанные с приложением, так как на графиках мы видели очень резкие снижения заказов и регистрации новых пользователей начиная с октября 2015. Стоит узнать причины данных падений.
- Для более корректных результатов стоит также сравнить retention между когортами зарегистрированных в июле 2015 и предыдущих годов (то есть за аналогичный период прошлых лет), но в исходных данных у нас отсутствуют наблюдения о заказах ранее 2015 года
